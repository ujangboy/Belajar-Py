```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
import scipy.stats as stats
from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.stats.stattools import durbin_watson
import matplotlib.pyplot as plt
```

```python
# Baca file Excel
df = pd.read_excel('c:\Users\a7916\Downloads\data_asumsi klasik.xlsx')

# Simpan sebagai CSV
df.to_csv('data_asumsi_klasik.csv', index=False)
print("Konversi selesai. File 'data_asumsi_klasik.csv' berhasil dibuat.")
```

```python
df = pd.read_csv('data_asumsi_klasik.csv')
print(" 5 Baris Pertama Data:")
display(df.head())

print("\n Statistik Deskriptif:")
display(df.describe())

# Cek missing value
print(f"\n Missing Value per kolom:\n{df.isnull().sum()}")
# Pisahkan variabel
X = df[['PER', 'ROI']]
y = df['Price']

# Tambahkan konstanta (intercept)
X = sm.add_constant(X)

# Fit model OLS
model = sm.OLS(y, X).fit()

# Tampilkan ringkasan regresi
print(model.summary())
```

```python


print("="*50)
print("🔍 UJI ASUMSI KLASIK REGRESI")
print("="*50)

# 1. MULTIKOLINEARITAS (VIF)
print("\n🔹 1. VIF (Multikolinearitas):")
vif_data = pd.DataFrame()
vif_data["Variabel"] = X.columns
vif_data["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
print(vif_data)
print("   → Aturan: VIF < 10 (idealnya < 5) berarti aman.")

# 2. NORMALITAS RESIDUAL (Shapiro-Wilk & Q-Q Plot)
resid = model.resid
stat, p_shapiro = stats.shapiro(resid)
print(f"\n🔹 2. Normalitas (Shapiro-Wilk):")
print(f"   Statistik = {stat:.4f}, p-value = {p_shapiro:.4f}")
print(f"   → H0: Data residual berdistribusi normal. Jika p > 0.05, gagal tolak H0 (normal).")

plt.figure(figsize=(5,4))
sm.qqplot(resid, line='45')
plt.title("Q-Q Plot Residual")
plt.show()

# 3. HETEROSKEDASTISITAS (Breusch-Pagan)
bp_stat, bp_pvalue, _, _ = sms.het_breuschpagan(resid, model.model.exog)
print(f"\n🔹 3. Heteroskedastisitas (Breusch-Pagan):")
print(f"   p-value = {bp_pvalue:.4f}")
print(f"   → H0: Homoskedastisitas (varians residual konstan). Jika p > 0.05, asumsi terpenuhi.")

# Plot Residual vs Fitted
plt.figure(figsize=(6,4))
plt.scatter(model.fittedvalues, resid, alpha=0.7)
plt.axhline(0, color='red', linestyle='--')
plt.xlabel("Fitted Values")
plt.ylabel("Residuals")
plt.title("Residual vs Fitted")
plt.show()

# 4. AUTOKORELASI (Durbin-Watson)
dw = durbin_watson(resid)
print(f"\n🔹 4. Autokorelasi (Durbin-Watson):")
print(f"   Nilai DW = {dw:.4f}")
print(f"   → Ideal: mendekati 2. <1.5 atau >2.5 mengindikasikan masalah autokorelasi.")

# 5. LINEARITAS (Secara visual & statistik)
print("\n🔹 5. Linearitas:")
print("   → Terlihat dari scatter plot X vs Y dan pola residual yang acak di sekitar 0.")
```
